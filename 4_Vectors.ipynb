# Vectors  Ploting

[matplotlib documentation quiver](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.quiver.html)

**`matplotlib.pyplot.quiver(*args, data=None, **kw)`**

Plot a 2-D field of arrows.

Call signatures:

**`quiver(U, V, **kw)`**

**`quiver(U, V, C, **kw)`**

**`quiver(X, Y, U, V, **kw)`**

**`quiver(X, Y, U, V, C, **kw)`**

**U **and **V **are the arrow data, **X** and **Y** set the location of the arrows, and** C** sets the color of the arrows. These arguments may be **1-D** or **2-D **arrays or sequences.

If X and Y are absent, they will be generated as a uniform grid. If U and V are 2-D arrays and X and Y are 1-D, and if len(X) and len(Y) match the column and row dimensions of U, then X and Y will be expanded with **`numpy.meshgrid()`**.


```
plt.quiver (0, 0, 3, 4)
plt.show()
```

```
plt.quiver (0, 0, 3, 4, scale_units='xy', angles='xy', scale = 1)
plt.show()
```

```
plt.quiver (0, 0, 3, 4, scale_units='xy', angles='xy', scale = 1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```
plt.quiver (0, 0, 3, 4, scale_units='xy', angles='xy', scale = 1)
plt.quiver (0, 0, -3, 4, scale_units='xy', angles='xy', scale = 1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```
plt.quiver (0, 0, 3, 4, scale_units='xy', angles='xy', scale = 1, color='r')
plt.quiver (0, 0, -3, 4, scale_units='xy', angles='xy', scale = 1, color='g')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```
# writiing a function ...which given list of these of vectros plot this in one go
# plot_vectors-- it suppose to take the list of vectors ---& for each vector we want these these 4 numbers
# so what a good data set for  that--->we can use tuples ---tuples are good for orderes collection of fix number of items which are immutable

def plot_vectors(vecs):
  colors = ['r', 'b', 'g', 'y']
  i = 0
  for vec in vecs:
    plt.quiver(vec[0], vec[1], vec[2], vec[3], scale_units='xy', angles='xy', scale = 1, color=colors[i%len(colors)]) # something would be cycloing between these
    i += 1                                                                                                            # as 'i' keep on incrementing it will go throw this one by one
  plt.xlim(-10, 10)                                                                                                   # & this % takes reminder so that this always remain 0 and len of colors
  plt.ylim(-10, 10)
  plt.show() 
   
```

```
plot_vectors([(0, 0, 3, 4)]) #check the braces for the data to be the form of a tuple
```

```
plot_vectors([(0, 0, 3, 4), (0, 0, -3, 4), (0, 0, -3, -2), (0, 0, 4, -1)]) 
```

```
a = np.asarray([3, 4])
b = np.asarray([4, -1])
```

$\vec{a}.\vec{b}==|\vec{a}||\vec{b}|cos(\theta)=a_xb_x+a_yb_y$



```
a_dot_b = np.dot(a, b)
```

```
print(a_dot_b)
```

**projection of a on b**

we draw a perpendicular from **a** onto **b** and we want to the  see the size of **a**



$a_b=|\vec{a}|\cos(\theta) =|\vec{a}| \frac{\vec{a}\cdot\vec{b}}{|\vec{a}||\vec{b}|} = \frac{\vec{a}\cdot\vec{b}}{|\vec{b}|} $

```
a_b = np.dot(a, b)/np.linalg.norm(b)
```

```
print(a_b)
```

*   In order to plot the vector we need both magnitude and  direction
*   so the direction of 'a_b' it is in the same direction as b 
*   so to find the vector along direcion of any vector we need UNIT vector alog that direction

* $\vec{a_b} = a_b\hat{b} = a_b\frac{\vec{b}}{|\vec{b}|}$







```
vec_a_b = (a_b/np.linalg.norm(b))*b
print(vec_a_b)

# so now we have a vector with some direction given by the two co-ordinates
```

# Linear combination

$\vec{c} = w_1 \vec{a} + w_2 \vec{b}$

```
def plot_linear_combination(a, b, w1, w2):
  
  plt.quiver(0,0,a[0],a[1], scale_units='xy', angles='xy', scale=1, color='r')
  plt.quiver(0,0,b[0],b[1], scale_units='xy', angles='xy', scale=1, color='b')
  
  c = w1 * a + w2 * b
  
  plt.quiver(0,0,c[0],c[1], scale_units='xy', angles='xy', scale=1, color='g')

  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

```
a = np.asarray([3, 4])
b = np.asarray([1.5, 2])
```

```
plot_linear_combination(a, b, -1, 1)
```

```
def plot_span(a, b):
  
  for i in range(1000):
    w1 = (np.random.random(1) - 0.5) * 3
    w2 = (np.random.random(1) - 0.5) * 3
    c = w1 * a + w2 * b
    plt.quiver(0,0,c[0],c[1], scale_units='xy', angles='xy', scale=1, color='g')
    
  plt.quiver(0,0,a[0],a[1], scale_units='xy', angles='xy', scale=1, color='r')
  plt.quiver(0,0,b[0],b[1], scale_units='xy', angles='xy', scale=1, color='b')

  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
```

```
plot_span(a, b)
```
